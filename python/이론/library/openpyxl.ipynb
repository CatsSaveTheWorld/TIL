# Openpyxl

```python
from openpyxl import Workbook

# 엑셀파일 쓰기
ws = Workbook()

# 이름이 있는 시트를 생성
ws1 = ws.create_sheet('시트1')
ws2 = ws.create_sheet('시트2', 1)

# 시트1에다 입력
ws1 = ws.active
ws1['A1'] = '숫자'

# 시트1 행 단위로 추가
ws1.append([1, 2, 3, 4, 110])

# 시트2에다 입력
ws2['A1'] = '문자'

# 시트2 행 단위로 추가
ws2.append(['가', '나', '다', '라', '마'])

#셀 단위로 추가
ws1.cell(5, 5, '5행 5열')   # 셀 단위는 0이 아닌 1부터 시작
ws2.cell(3, 1, '3행 1열')
ws.save("./data/test.xlsx")
```
